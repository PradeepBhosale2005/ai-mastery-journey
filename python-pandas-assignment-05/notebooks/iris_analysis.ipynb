# Iris Dataset Analysis with Pandas

This notebook demonstrates dataset cleaning, summary statistics, and visual analysis using Pandas and Matplotlib.

Dataset used: `data/iris_sample.csv`


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load dataset from a robust path.
current_dir = Path.cwd()
data_path = current_dir / "data" / "iris_sample.csv"

if not data_path.exists():
    data_path = current_dir.parent / "data" / "iris_sample.csv"

df = pd.read_csv(data_path)
df.head()
```

## 1. Data Cleaning

Cleaning steps:
- Check missing values
- Fill missing numeric values with column means
- Remove duplicate rows


```python
print("Original shape:", df.shape)
print("\nMissing values before cleaning:")
print(df.isna().sum())

numeric_columns = df.select_dtypes(include="number").columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

df_cleaned = df.drop_duplicates()

print("\nCleaned shape:", df_cleaned.shape)
print("\nMissing values after cleaning:")
print(df_cleaned.isna().sum())

df_cleaned.head()
```

## 2. Summary Statistics


```python
df_cleaned.describe()
```

```python
df_cleaned.groupby("species")[numeric_columns].mean()
```

## 3. Plot 1: Species Count


```python
species_counts = df_cleaned["species"].value_counts()

plt.figure(figsize=(6, 4))
species_counts.plot(kind="bar")
plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

## 4. Plot 2: Sepal Length Distribution


```python
plt.figure(figsize=(6, 4))
df_cleaned["sepal_length"].plot(kind="hist", bins=8)
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## 5. Plot 3: Sepal Length vs Petal Length


```python
plt.figure(figsize=(6, 4))

for species, group in df_cleaned.groupby("species"):
    plt.scatter(group["sepal_length"], group["petal_length"], label=species)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.legend()
plt.tight_layout()
plt.show()
```

## 6. Plot 4: Sepal Width by Species


```python
plt.figure(figsize=(6, 4))
df_cleaned.boxplot(column="sepal_width", by="species")
plt.title("Sepal Width by Species")
plt.suptitle("")
plt.xlabel("Species")
plt.ylabel("Sepal Width")
plt.tight_layout()
plt.show()
```
